# Flight Delay Project — Cleaning and EDA

In this notebook, we clean the raw flight data and prepare it for analysis.

We will:
- keep only the columns needed for our project
- remove cancelled flights from the main analysis dataset
- handle missing values in delay-related columns
- convert date columns to proper datetime format
- save a cleaned version of the dataset for later use

```python
from pathlib import Path
import pandas as pd

# go ONE level up (not parent.parent)
project_root = Path().resolve().parent

data_path = project_root / "data" / "raw" / "flights_sample_3m.csv"

print(data_path)  # verify path

df = pd.read_csv(data_path)
```

```python
print(data_path.exists())
```

```python
columns_needed = [
    'FL_DATE',
    'AIRLINE',
    'ORIGIN',
    'DEST',
    'DEP_DELAY',
    'ARR_DELAY',
    'CANCELLED',
    'DIVERTED',
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

df = df[columns_needed]
```

```python
df['CANCELLED'].value_counts()
```

```python
df = df[df['CANCELLED'] == 0]
```

```python
df = df.dropna(subset=['ARR_DELAY', 'DEP_DELAY'])
```

```python
df = df.dropna(subset=['ARR_DELAY', 'DEP_DELAY'])
```

```python
delay_cols = [
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

df[delay_cols] = df[delay_cols].fillna(0)
```

```python
df.shape
df.isna().sum()
```

```python
output_path = project_root / "data" / "processed" / "flights_cleaned.csv"
df.to_csv(output_path, index=False)
```
